**Python Operators**

```python
#Aithmetic operators
#+,-,*,/,//,%,**
print(45+36)
print(45-36)
print(45*36)

print(10/3)  #Division
print(10//3)  #Floor division:returns on int part
print(10%3)  #Modulo/remainder  operator

print(7**3)  #x^y
```

```python
#Assignment operators =,+=,-=,*=,/=
age=45
print(age)

age+=10 #age=age+10
print(age)

age-=5 #age=age-5
print(age)
```

```python
#Relational/Conditional Operators
#<,>,<=,>=,==,!=

var1=56
var2=25
print(var1==var2) #Equality Operator
print(var1!=var2) #not equal

print(var1>var2)
print(var1<var2)
print(var1>=var2)
print(var1<=var2)
```

```python
#Logical Operators: and,or,not
age=45
print(age>18 and age<=40)
print(age>18 or age<=40)
print(age>60 or (age>18 and age<40))
```
