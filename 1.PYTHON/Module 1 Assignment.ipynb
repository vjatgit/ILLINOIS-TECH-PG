```python
# 1. Write a program to find the factors of agiven number and 
#    check whether the factor is even or odd.

num = int(input("Enter a number: "))
print(f"Factors of {num} and their even/odd status:")
for i in range(1, num + 1):
    if num % i == 0:
        if i % 2 == 0:
            print(f"{i} is Even")
        else:
            print(f"{i} is Odd")

# ANSWER
# Enter a number:  102
# Factors of 102 and their even/odd status:
# 1 is Odd
# 2 is Even
# 3 is Odd
# 6 is Even
# 17 is Odd
# 34 is Even
# 51 is Odd
# 102 is Even
```

```python
#===================================================================
# 2. Write a code that accepts a sequence of words as input and prints
#    the words in a sequence after sorting them alphabetically.

words = input("\nEnter a sequence of words: ")
word_list = words.split()
word_list.sort()
print("Sorted words:")
for word in word_list:
    print(word)

# ANSWER
#   Enter a sequence of words:  ab cd ef z xy uvw
#Sorted words:
# ab
# cd
# ef
# uvw
# xy
# z  
```

```python
# 3. Write a program that will find all the numbers between 1000 and 3000
#    (both excluded) such that each digit of a number is an odd number.
#    Print the number of such elements

result = []
for num in range(1001, 3000):
    num_str = str(num)
    if all(int(digit) % 2 != 0 for digit in num_str):
        result.append(num)
print("\nNumbers with all odd digits:", result)
print("Total count:", len(result))


#ANSWER:
# Numbers with all odd digits: [1111, 1113, 1115, 1117, 1119, 1131, 1133, 1135, 1137, 1139, 1151, 1153, 1155, 
#1157, 1159, 1171, 1173, 1175, 1177, 1179, 1191, 1193, 1195, 1197, 1199, 1311, 1313, 1315, 1317, 1319, 1331, 
#1333, 1335, 1337, 1339, 1351, 1353, 1355, 1357, 1359, 1371, 1373, 1375, 1377, 1379, 1391, 1393, 1395, 1397, 
#1399, 1511, 1513, 1515, 1517, 1519, 1531, 1533, 1535, 1537, 1539, 1551, 1553, 1555, 1557, 1559, 1571, 1573, 
#1575, 1577, 1579, 1591, 1593, 1595, 1597, 1599, 1711, 1713, 1715, 1717, 1719, 1731, 1733, 1735, 1737, 1739, 
#1751, 1753, 1755, 1757, 1759, 1771, 1773, 1775, 1777, 1779, 1791, 1793, 1795, 1797, 1799, 1911, 1913, 1915, 
#1917, 1919, 1931, 1933, 1935, 1937, 1939, 1951, 1953, 1955, 1957, 1959, 1971, 1973, 1975, 1977, 1979, 1991, 
#1993, 1995, 1997, 1999]
#Total count: 125
```

```python
# 4. Write a program that accepts a stringand calculatesthe number of letters
#    and digits.Suppose if the entered string is: Edureka123
#    Then the output will be:LETTERS: -7DIGITS: -3

input_str = input("\nEnter a string: ")
letters = 0
digits = 0
for ch in input_str:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1
print("LETTERS:", letters)
print("DIGITS:", digits)

#ANSWER
# Enter a string:  Edureka 123
#LETTERS: 7
#DIGITS: 3
```

```python
# 5. Design a code which will find whether
#    the given number is Palindrome number or not.

num = input("\nEnter a number to check if it is a palindrome: ")
if num == num[::-1]:
    print(f"{num} is a Palindrome")
else:
    print(f"{num} is NOT a Palindrome")

#ANSWER ( This question was confusing to understand. Finally got the answer)
# Enter a number to check if it is a palindrome:  102
# 102 is NOT a Palindrome

# Enter a number to check if it is a palindrome:  1002001
# 1002001 is a Palindrome

# Enter a number to check if it is a palindrome:  202
# 202 is a Palindrome
```
